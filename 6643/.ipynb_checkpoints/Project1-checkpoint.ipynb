```python
import numpy as np
np.random.seed(0) # set fixed seed
```

```python
# Part I: Perform the numerical experiments for random square matrices.
m = 10 # dimension of matrix
mu = 0 # mean is 0
sigma = np.sqrt(m) # standard deviation is square root of m 
```

```python
# (1) Compute the 2-norm for the random matrices and find out how it behaves as m → ∞.
```

```python
# (2) Compute the eigenvalues of the random matrices
#     plot them and find out patterns if there is any. 
#     You need to do it for m = 16,32,64,128,256,···. 
#     What can you say about the spectral radius as m → ∞.
m_vec = map(lambda n: 2**n, range(4,11))
for m in m_vec:
    random_matrix = np.random.normal(mu, sigma, m**2); random_matrix.shape = (m,m)
    print np.linalg.eigvals(random_matrix)
```
